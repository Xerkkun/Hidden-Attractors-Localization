# hidden-attractors-fo quickstart

This notebook shows the public package workflow: import the model, inspect equilibria, load reference candidates, and compute a lightweight trajectory diagnostic. It avoids long C/EFORK simulations so it can run as a documentation example.

## Imports

If the package is not installed, run `python -m pip install -e .` from the repository root first.

```python
import numpy as np

from hidden_attractors import chua_piecewise_parameters, load_final_candidate_records
from hidden_attractors.analysis import trajectory_metrics
from hidden_attractors.models import equilibria_piecewise, rhs_piecewise
```

## Check the Chua equilibria

A quick sanity check is to evaluate the vector field at every equilibrium.

```python
params = chua_piecewise_parameters()
for name, point in equilibria_piecewise(params).items():
    residual_norm = np.linalg.norm(rhs_piecewise(point, params))
    print(f"{name:>2}  point={point}  residual_norm={residual_norm:.3e}")
```

## Load reference candidates

The current examples load candidate metadata from the included reference `outputs/` folders.

```python
records = load_final_candidate_records()
for record in records:
    print(record.candidate_id, record.route, record.q, record.robust_start)
```

## Lightweight trajectory diagnostic

For a fast notebook example, use a synthetic bounded signal with the same column convention as workflow trajectories: `t, x, y, z`.

```python
t = np.linspace(0.0, 20.0, 2000)
trajectory = np.column_stack([
    t,
    np.sin(t),
    np.cos(1.7 * t),
    0.25 * np.sin(0.5 * t),
])

metrics, payload = trajectory_metrics(
    trajectory,
    h=float(t[1] - t[0]),
    t_start=5.0,
    divergence_norm=120.0,
    equilibrium_tol=1.0e-3,
)

{key: metrics[key] for key in ["bounded", "range_x", "range_y", "range_z", "fft_peak", "section_points"]}
```
